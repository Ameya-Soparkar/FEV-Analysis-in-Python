```python
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
import statsmodels.api as sm
df= pd.read_csv('fev.txt', usecols=['AGE', 'FEV', 'HEIGHT','SEX', 'SMOKE'])
df['SMOKE'] = df['SMOKE'].replace({2:0})
df['SEX'] = df['SEX'].replace({2:0})
```

```python
def statss_multivariate(model, *args):
    a=model.summary()
    results_as_html = a.tables[0].as_html()
    d=pd.read_html(results_as_html, header=None)[0]
    values = d.iloc[0:6,2:4].reset_index(drop=True)
    values.columns = ['Stats', 'Model']
    results_as_html_1 = a.tables[1].as_html()
    c=pd.read_html(results_as_html_1, header=None)[0]
    c=c.transpose()
    c.columns= args           # 'Stats', 'Const', 'X variables in order arranged in model'
    c.drop(c.head(1).index, inplace=True)
    ddf= pd.concat([c, values],ignore_index=True)
    ddf.fillna(0, inplace=True)
    return ddf



def statss_bivariate(model,*args):
    a=model.summary()
    results_as_html = a.tables[0].as_html()
    d=pd.read_html(results_as_html, header=None)[0]
    values = d.iloc[0:6,2:4].reset_index(drop=True)
    values.columns = ['Stats', 'Model']
    results_as_html_1 = a.tables[1].as_html()
    c=pd.read_html(results_as_html_1, header=None)[0]
    c=c.transpose()
    c.columns= args # 'Stats', 'Const', 'X variable'
    c.drop(c.head(1).index, inplace=True)
    ddf= pd.concat([c, values],ignore_index=True)
    ddf.fillna(0, inplace=True)
    return ddf
```
